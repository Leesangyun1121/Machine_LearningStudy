# data Classification

```python
#k평균 알고리즘(데이터 군집화기술)

from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(columns = ['x','y'])
```

```python
import random
for i in range(30):
    df.loc[i] = [random.randrange(0,20),random.randrange(0,20)]
    print(df.loc[i])
```

```python
df.head(30)
```

```python
sb.lmplot('x','y',data = df, fit_reg = False ,scatter_kws = {"s":100})
plt.title('KMeans example')

plt.xlabel('x')
plt.ylabel('y')
```

```python
points = df.values
kmeans = KMeans(n_clusters=4).fit(points)
print(kmeans.cluster_centers_)
print(kmeans.labels_)
```

```python
df['cluster'] = kmeans.labels_
df.head(30)
```

```python
sb.lmplot('x','y',data = df, fit_reg = False ,scatter_kws = {"s":150},hue = 'cluster')
plt.title('KMeans example')
```

## 평균,최대,최소 온도 및 강수량을 데이터로 배추값 예측하기 실습

```python
from pandas.io.parsers import read_csv

model = tf.global_variables_initializer()
data = read_csv('price data.csv',sep=',')

xy = np.array(data,dtype = np.float32)
print(xy)
```

```python
x_data = xy[:,1:-1]
y_data = xy[:,[-1]]

X = tf.placeholder(tf.float32,shape=[None,4])
Y = tf.placeholder(tf.float32,shape=[None,1])
W = tf.Variable(tf.random_normal([4,1]),name ='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')
print(W)
```

```python
print(x_data)
```

```python
hypo = tf.matmul(X,W) + b #matmul은 행렬곱
cost = tf.reduce_mean(tf.square(hypo - Y))
optimizer= tf.train.GradientDescentOptimizer(learning_rate = 0.000005)
train = optimizer.minimize(cost)
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(100001):
    cost_,hypo_,_ = sess.run([cost,hypo,train],feed_dict={X:x_data,Y:y_data})
    if step % 500 ==0:
        print("#",step," 손실비용",cost_)
        print("배추가격",hypo_[0])
```

```python
saver = tf.train.Saver()
save_path = saver.save(sess,"./saved.cpkt")
```
